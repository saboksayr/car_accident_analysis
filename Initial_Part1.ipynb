## 1. The following are the histograms of car accidents in the Unted States and three states with the highest rate of car accidents.

```python
plt.figure(figsize=(14,15))

plt.subplot(2,2,1)
cities_US = df["City"].value_counts().sort_values(ascending=False)
cities_US.head(30).plot(kind='bar', color = 'green')
plt.title('Accident in the United States', weight='bold', size = 12)
plt.xlabel('City', weight = 'bold', size = 12)
plt.ylabel('Frequency', weight = 'bold', size =12)

plt.subplot(2,2,2)
df_CA = df[df['State'] == 'CA']
cities_CA = df_CA["City"].value_counts().sort_values(ascending=False)
cities_CA.head(20).plot(kind='bar', color = 'green')
plt.title('Accident in California', weight='bold', size = 12)
plt.xlabel('City', weight = 'bold', size = 12)
plt.ylabel('Frequency', weight = 'bold', size =12)

plt.subplot(2,2,3)
df_TX = df[df['State'] == 'TX']
cities_TX = df_TX["City"].value_counts().sort_values(ascending=False)
cities_TX.head(20).plot(kind='bar', color = 'green')
plt.title('Accident in Texas', weight='bold', size = 12)
plt.xlabel('City', weight = 'bold', size = 12)
plt.ylabel('Frequency', weight = 'bold', size =12)

plt.subplot(2,2,4)
df_SC = df[df['State'] == 'SC']
cities_SC = df_SC["City"].value_counts().sort_values(ascending=False)
cities_SC.head(20).plot(kind='bar', color = 'green')
plt.title('Accident in South Carolina', weight='bold', size = 12)
plt.xlabel('City', weight = 'bold', size = 12)
plt.ylabel('Frequency', weight = 'bold', size =12)

plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)
```

### The high rate of accident in a state does not necessarily mean that the rate of accident is high everywhere in the state. For instance in Texas, as we can see in the histogram, except a few cities, most of other cities have a low rate of car accident. However in some states such as California and South Carolina there are many cities with a considerable rate of accident.
